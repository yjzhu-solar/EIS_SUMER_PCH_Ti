```python
import matplotlib.pyplot as plt
from matplotlib import ticker
import numpy as np
import scipy.io 
import sys
sys.path.append('/Users/yjzhu/Desktop/Solar/MyPy')
import juanfit
import importlib
importlib.reload(juanfit)
from juanfit import SpectrumFitSingle
import cmcrameri.cm as cmcm
```

```python
spectrum_sav = scipy.io.readsav("../../save/spectrum/noti/spectrum_pch_2063_sumer_window1_noti.sav")
syn_spec = spectrum_sav["w"][:,0,:]
wvl = spectrum_sav["x"][0,0,0,:]
y_coord = spectrum_sav["x"][2,:,0,0]
```

```python
y_coord[6:10]
```

```python
fig, ax = plt.subplots(figsize=(16,5))
ax.pcolormesh(np.arange(syn_spec.shape[1]),y_coord,np.nan_to_num(np.log10(syn_spec),nan=-np.inf),
                  cmap=cmcm.lajolla_r,shading="auto",vmin=-5)

ax_wvl = ax.twiny()
ax_wvl.set_xlim(wvl[0],wvl[-1])
ax.xaxis.set_minor_locator(ticker.AutoMinorLocator(10))
ax_wvl.xaxis.set_minor_locator(ticker.AutoMinorLocator(10))
for ax_ in (ax, ax_wvl):
    ax_.tick_params(labelsize=16)
    ax_.tick_params(which="major",width=1.2,length=8)
    ax_.tick_params(which="minor",width=1.2,length=4)
```

```python
FeXI_680_slice = (200,250)
wvl_FeXI_680 = wvl[FeXI_680_slice[0]:FeXI_680_slice[1]]
int_FeXI_680 = np.average(syn_spec[6:10,FeXI_680_slice[0]:FeXI_680_slice[1]],axis=0)

FeXI_680_fit_model = SpectrumFitSingle(data=int_FeXI_680,wvl=wvl_FeXI_680,err=None,
                                      same_width=False,line_number=2,line_wvl_init=[679.7,680.4],
                                      int_max_init=[1,5],fwhm_init=[0.2,0.2])
#FeXI_680_fit_model.plot(plot_fit=False,plot_params=False)
FeXI_680_fit_model.run_lse()
FeXI_680_fit_model.plot(plot_title=r"\textbf{Fe \textsc{xi} 680 \mbox{\AA}}",
                       xlabel=r"Wavelength $\lambda$ [\mbox{\AA}]",
                        ylabel=r"Intensity $\mathrm{[erg \cdot s^{-1} \cdot cm^{-2} \cdot \AA^{-1} sr^{-1}]}$",
                       color_style="Cold")
```
