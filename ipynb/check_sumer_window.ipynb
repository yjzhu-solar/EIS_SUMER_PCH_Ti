```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
plt.style.use("science")
from glob import glob
```

```python
#338 
filenames_338 = glob("../raw_data/sumer/20071116/level1/sum_r_20071116_09*0338_l1.sav")
filenames_338 = sorted(filenames_338)

sumer_img_338_0 = scipy.io.readsav(filenames_338[0])["image_lvl1"]
sumer_img_338_1 = scipy.io.readsav(filenames_338[1])["image_lvl1"]
sumer_img_338_2 = scipy.io.readsav(filenames_338[2])["image_lvl1"]
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.step(np.arange(360),np.sum(sumer_img_338_0,axis=1),where="mid")
ax.tick_params(labelsize=16)
ax.axvspan(6,294,color="red",alpha=0.3)
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.step(np.arange(360),np.sum(sumer_img_338_1,axis=1),where="mid")
ax.tick_params(labelsize=16)
ax.axvspan(6,294,color="red",alpha=0.3)
```

```python
fig, ax = plt.subplots(figsize=(16,6))
ax.step(np.arange(360),np.sum(sumer_img_338_2,axis=1),where="mid")
ax.tick_params(labelsize=16)
ax.axvspan(6,294,color="red",alpha=0.3)
```
