```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.io
plt.style.use("science")
from juanfit import SpectrumFitSingle
```

```python
spectrum_save = scipy.io.readsav("/home/yjzhu/SOLAR/SWMF_20210924/SWMF/GM/BATSRUS/srcPostProc/spectrum_pch_2063_neviii.sav",verbose=True)
```

```python
spectrum_ti_save = scipy.io.readsav("/home/yjzhu/SOLAR/SWMF_20210924/SWMF/GM/BATSRUS/srcPostProc/spectrum_pch_2063_neviii_ti.sav",verbose=False)
```

```python
spectrum = spectrum_save["w"][:,0,:]
wvl = spectrum_save["x"][0,0,0,:]
spectrum_ti = spectrum_ti_save["w"][:,0,:]
wvl_ti = spectrum_ti_save["x"][0,0,0,:]
```

```python
fig, ax = plt.subplots(figsize=(8,5))
ax.step(wvl,spectrum[12,:],where="mid",lw=2)
ax.step(wvl_ti,spectrum_ti[12,:],where="mid",lw=2)
ax.set_xlim(768,772)
ax.tick_params(labelsize=16)
ax.set_xlabel("$\lambda$ [\mbox{\AA}]",fontsize=16)
```

```python
np.argmin(np.abs(wvl-772))
```

```python
fit_model = SpectrumFitSingle(data=spectrum[12,90:160],wvl=wvl[90:160],
                             line_number=1,line_wvl_init=[770.4],int_max_init=25,
                             fwhm_init=[0.2])
fit_model.run_lse()
fit_model.plot()
```

```python
fit_model = SpectrumFitSingle(data=spectrum_ti[12,90:160],wvl=wvl_ti[90:160],
                             line_number=1,line_wvl_init=[770.4],int_max_init=25,
                             fwhm_init=[0.2])
fit_model.run_lse()
fit_model.plot()
fit_model.int_total_fit
```
