```python
fig = plt.figure(figsize=(18,9),constrained_layout=False)
for ii in range(8):
    gs_left = (ii%4)*0.24 + 0.06
    gs_right = (ii%4)*0.24 + 0.26
    gs_top = 0.99 - (np.floor_divide(ii,4))*0.47
    gs_bottom = 0.55 - (np.floor_divide(ii,4))*0.47
    if ii < 4:
        gs = fig.add_gridspec(nrows=9, ncols=1, left=gs_left, right=gs_right,bottom=gs_bottom,top=gs_top,
                        hspace=0.)
    else:
        gs = fig.add_gridspec(nrows=9, ncols=1, left=gs_left, right=gs_right,top=gs_top,
                        hspace=0.)
    ax_fit = fig.add_subplot(gs[:6, :])
    ax_res = fig.add_subplot(gs[6:-1, :])


    if (ii == 0) or (ii == 4):
         #ax_fit.set_ylabel(r"$\boldsymbol{I_\lambda}$\,\textbf{[\si{erg.s^{-1}.cm^{-2}.\angstrom^{-1}.sr^{-1}}]}",
         #fontsize=18)
        ax_fit.set_ylabel(r"$\boldsymbol{I_\lambda}$",
        fontsize=18)
        ax_res.set_ylabel(r"$\boldsymbol{r}$",
        fontsize=18)
    
    if ii > 3:
        ax_res.set_xlabel(r"$\lambda$\textbf{\mbox{\AA}}",fontsize=18)
```
